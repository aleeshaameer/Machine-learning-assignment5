```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd

# Load the Iris dataset
iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Drop the species column if it exists (it shouldn't in this case)
# But including this for safety and generalization to other datasets
if 'species' in iris_df.columns:
    X = iris_df.drop(columns=['species'])
else:
    X = iris_df  # Use the entire DataFrame if 'species' column not found

# Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=0)
# Fit the KMeans model to your data
kmeans.fit(X) #This line is added to fit the kmeans model to the data 'X'
# Get the cluster labels
labels = kmeans.labels_

# Visualize the clusters
plt.scatter(X['sepal length (cm)'], X['sepal width (cm)'], c=labels)
plt.title('KMeans Clustering')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
import numpy as np
import matplotlib.pyplot as plt

# Apply Hierarchical clustering
agg_clustering = AgglomerativeClustering(n_clusters=3, linkage='ward')
agg_clustering.fit(X)
labels = agg_clustering.labels_

# Visualize the clusters
plt.scatter(X['sepal length (cm)'], X['sepal width (cm)'], c=labels)
plt.title('Hierarchical Clustering')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
```
